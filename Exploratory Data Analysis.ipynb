# Univariate Analysis for categorical Data

```python
df['Location'].value_counts().plot.bar(title='Freq dist of Location')
```

```python
label=list(x.index)
count=x.Price
plt.bar(label,count)
plt.show()
```

```python
label=list(x.index)
count=x.Price
plt.pie(count,labels=label,autopct='%1.2f%%')
plt.show()
```

```python
col_names=['Price','Mileage(kmpl)']
fig, ax = plt.subplots(len(col_names), figsize=(10,12))

for i, col_val in enumerate(col_names):
    sns.distplot(df[col_val], hist=True, ax=ax[i])
    ax[i].set_title('Freq dist '+col_val, fontsize=10)
    ax[i].set_xlabel(col_val, fontsize=8)
    ax[i].set_ylabel('Count', fontsize=8)
plt.show()
```

```python
a=df['Price'].dropna()
sns.distplot(a)         
plt.show()
```

# Correlation Matrix
A correlation matrix is a table showing the value of the correlation coefficient (Correlation coefficients are used in statistics to measure how strong a relationship is between two variables. ) between sets of variables. Each attribute of the dataset is compared with the other attributes to find out the correlation coefficient.

```python
f, ax = plt.subplots(figsize=(10, 8))
corr = data.corr()
sns.heatmap(corr,xticklabels=corr.columns.values, yticklabels=corr.columns.values,annot=True)
```
